```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler

# Step 1: Read the file
data = pd.read_csv('Electricity Generated.csv', header=None)

# Step 2: Preprocess the data
# Assuming the first column is time and the second column is electricity generated
data.columns = ['Time', 'Electricity Generated']
data['Time'] = pd.to_datetime(data['Time'], format='%y-%b')
data['Electricity Generated'] = data['Electricity Generated'].str.replace(',', '').astype(float)
data.set_index('Time', inplace=True)

# Step 3: Prepare the data for training
X = data.index.to_julian_date().values.reshape(-1, 1)
y = data['Electricity Generated'].values.reshape(-1, 1)

# Step 4: Train the neural network
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

model = MLPRegressor(hidden_layer_sizes=(100, 100), activation='relu', random_state=42)
model.fit(X_scaled, y_scaled)

# Step 5: Generate forecasts
forecast_dates = pd.date_range(start='2023-04-01', end='2030-12-01', freq='MS')
forecast_dates_julian = forecast_dates.to_julian_date().values.reshape(-1, 1)
forecast_dates_scaled = scaler_X.transform(forecast_dates_julian)
forecast_scaled = model.predict(forecast_dates_scaled.reshape(-1, 1))
forecast = scaler_y.inverse_transform(forecast_scaled.reshape(-1, 1))

# Step 6: Plot the data and forecast
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Electricity Generated'], label='Actual')
plt.plot(forecast_dates, forecast, '--', label='Forecast')
plt.xlabel('Time')
plt.ylabel('Electricity Generated (Gwh)')
plt.title('Electricity Generation Forecast')
plt.legend()
plt.grid(True)
plt.show()

```
