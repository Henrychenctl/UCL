```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os

# Data
years = np.array([2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021])
overall = np.array([35489.3, 36801.8, 38304.9, 38986.9, 38822.9, 42251.8, 43007.1, 44200.7, 44948.8, 46403.0, 47513.9, 48623.0, 49637.9, 50460.9, 51730.1, 50779.6, 53483.5])

# Prepare training data
X = years.reshape(-1, 1)
y = overall

# Train the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)

# Predict for 2022-2030
future_years = np.arange(2022, 2031).reshape(-1, 1)
predicted_overall = model.predict(future_years)

# Step 6: Plot the data and forecast
plt.figure(figsize=(12, 6))
plt.plot(years, overall, label='Actual', color='b')
plt.plot(future_years, predicted_overall, '--', label='Forecast', color='r')
plt.xlabel('Year')
plt.ylabel('Overall')
plt.title('Overall Trend (2005-2021) with Prediction (2022-2030)')
plt.legend()
plt.grid(True)

# Customize plot style
plt.style.use('ggplot')
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)

# Save the plot to desktop
output_path = os.path.expanduser("~/Desktop/overall_trend_and_forecast.png")
plt.savefig(output_path, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os

data = pd.read_csv('Overall.csv')
overall_data = data.iloc[0, 1:].values

# Step 2: Preprocess the data
overall_data_processed = np.array([float(value.replace(',', '')) for value in overall_data])
scaler = MinMaxScaler()
overall_data_scaled = scaler.fit_transform(overall_data_processed.reshape(-1, 1))

# Split the data into input (X) and target (y) sequences
sequence_length = 5  # Adjust as needed
X_train = []
y_train = []
for i in range(sequence_length, len(overall_data_scaled)):
    X_train.append(overall_data_scaled[i-sequence_length:i, 0])
    y_train.append(overall_data_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)

# Reshape X_train to match the expected input shape of LSTM
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))

# Step 3: Train a neural network model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=100, batch_size=32)

# Step 4: Generate predictions for 2005 to 2030
past_years = range(2005, 2023)
future_years = range(2023, 2031)

X_past = overall_data_processed[:len(past_years)]
X_future = overall_data_processed[-sequence_length:]
X_past_scaled = scaler.transform(X_past.reshape(-1, 1))
X_future_scaled = scaler.transform(X_future.reshape(-1, 1))

X_past_scaled = X_past_scaled.reshape(1, 1, len(past_years)-1)
X_future_scaled = X_future_scaled.reshape(1, 1, sequence_length)

y_pred_scaled = model.predict(X_future_scaled)
y_pred = scaler.inverse_transform(y_pred_scaled)


# Step 5: Prepare the forecast data
forecast_dates_past = pd.date_range(start='2005', periods=len(past_years), freq='Y')
forecast_dates_future = pd.date_range(start='2023', periods=len(future_years), freq='Y')

# Update X_past with the forecasted values
X_past = np.concatenate([X_past, y_pred.flatten()])

# Update X_past with the forecasted values
X_past = np.concatenate([X_past, y_pred.flatten()])

# Step 6: Plot the data and forecast
plt.figure(figsize=(12, 6))
plt.plot(forecast_dates_past, X_past[:len(forecast_dates_past)], label='Actual', color='b')
plt.plot(forecast_dates_future[:-1], X_past[-len(forecast_dates_future[:-1]):], '--', label='Forecast', color='r')
plt.plot(forecast_dates_future[-1:], y_pred.flatten()[-1:], '--', color='r')
plt.xlabel('Time')
plt.ylabel('Overall Amount')
plt.title('Overall Amount Forecast')
plt.legend()
plt.grid(True)

# Customize plot style
plt.style.use('ggplot')
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)

# Save the plot to desktop

output_path = os.path.expanduser("~/Desktop/overall_amount_forecast.png")
plt.savefig(output_path, bbox_inches='tight')

# Show the plot
plt.show()


```
